```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 
dtype = None 
load_in_4bit = True 
```

```python
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "AJosh/lora_model1",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
```

```python
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
you are a workout plan generator,generate the workout plan according to the given input

### Input:
I am a {} -year-old {} with a height of {}  cm and a weight of {}  kg. My primary goal is {} , and I aim to reach a target weight of {} kg. I am a {}  at the gym. Can you create a 7-day workout plan for me?

### Response:
{}"""
```

```python
FastLanguageModel.for_inference(model)
def inference():
    age = int(input("Enter your age: "))
    sex = input("male or female:")
    height = (input("Enter your height in cm: "))
    weight = (input("Enter your current weight in kg: "))
    goal = input("Enter your primary goal (weight loss, endurance, muscle gain, health): ")
    target_weight = (input("Enter your target weight in kg: "))
    experience = input("Enter your experience level (beginner, intermediate, advanced): ")
    output = alpaca_prompt.format(age,sex,height,weight,goal,target_weight,experience," ")
    inputs = tokenizer([output], return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=700, use_cache=True)
    output_text = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
    print(output_text)
```
